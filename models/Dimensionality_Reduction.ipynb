## üß† Linear Discriminant Analysis (LDA)

üìå LDA, …ôn √ßox $C - 1$ √∂l√ß√ºl√º bir f…ôzaya ki√ßiltm…ô ed…ô bil…ôr ($C$: sinif sayƒ±).
Bu metod, x√ºsusil…ô t…ôsnifat (classification) probleml…ôrind…ô g√ºcl√ºd√ºr; lakin sinif paylanmasƒ± qeyri-b…ôrab…ôrdirs…ô v…ô ya veril…ôrd…ô sinifl…ôr x…ôtti olaraq ayrƒ±lmƒ±rsa t…ôsirli olmaya bil…ôr.

üìå ICA, PCA-dan f…ôrqli olaraq **y√ºks…ôk d…ôr…ôc…ôd…ôn statistik m√ºst…ôqillik** f…ôrz edir v…ô siqnal/≈ü…ôkil f…ôrql…ôndirm…ô kimi sah…ôl…ôrd…ô t…ôsirlidir. Lakin hesablama baxƒ±mƒ±ndan daha m√ºr…ôkk…ôbdir.
